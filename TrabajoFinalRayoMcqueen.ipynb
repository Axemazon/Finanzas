```python
# Sección 1: Preguntas de Aversión al Riesgo

# Obtener las respuestas del cliente
def obtener_respuestas():
    preguntas = [
        "1. ¿Cuál es el horizonte de tiempo en el que mantendrá su inversión?\n a. Máximo 3 años\n b. Entre 3 y 5 años\n c. Más de 5 años\n d. Más de 10 años",
        "2. ¿Cuál sería la máxima variación que estaría dispuesto(a) a tolerar sobre su inversión en un año para considerar liquidar su inversión?\n a. -5%\n b. -10%\n c. -15%\n d. No la cerraría, mantendría mi posición y esperaría a recuperarme.",
        "3. ¿En qué rango de edad o calidad se encuentra?\n a. Menor a 35 años\n b. Entre 35 y 50 años\n c. Mayor a 50 años\n d. Pensionado",
        "4. ¿Qué porcentaje del total de sus activos representa este ahorro o inversión?\n a. Hasta 10%\n b. Entre 10% y 30%\n c. Entre 30% y 50%\n d. Más del 50%",
        "5. Si tuviera la oportunidad de aumentar su rentabilidad asumiendo mayor riesgo e incluyendo pérdidas potenciales a su inversión inicial ¿Qué haría usted?\n a. No estaría dispuesto a asumir mayor riesgo\n b. Estaría dispuesto a asumir un poco más de riesgo tratando de no incurrir en pérdidas de capital\n c. Estaría dispuesto a asumir mucho más riesgo para aumentar mi rentabilidad\n d. Me endeudaría o buscaría opciones de apalancamiento para obtener un retorno más interesante",
        "6. ¿Cuáles son las expectativas de retorno para la inversión?\n a. No importa si la inversión sigue el ritmo de la inflación. El objetivo es preservar mi inversión (capital)\n b. Mi rentabilidad debería seguir el ritmo de la inflación con mínimo riesgo\n c. La rentabilidad debería ser un poco más que la inflación con un riesgo moderado (intermedio)\n d. La rentabilidad debería exceder significativamente la inflación incluso si esto pudiera significar altos niveles de riesgo",
        "7. ¿Cuál es el propósito de su ahorro o inversión?\n a. Acumular capital a mediano plazo con un fin específico (estudios, viajes, etc.)\n b. Ahorrar para pagar gastos del corto plazo\n c. Contar con una inversión que me permita tener una renta en la vejez\n d. Especular en el mercado de capitales aceptando las posibles pérdidas que tenga",
        "8. ¿Ha tenido experiencia en el tema de inversiones?\n a. Poca\n b. Alguna, preferiría recibir orientación y asesoría en los temas de inversiones\n c. Mucha, me siento seguro y tranquilo para entender los riesgos asociados a cada inversión y tomar decisiones\n d. Soy inversionista profesional"
    ]
    
    respuestas = []
    
    for pregunta in preguntas:
        print(pregunta)
        respuesta = input("Elija una opción (a, b, c, d): ").lower()
        while respuesta not in ['a', 'b', 'c', 'd']:
            print("Respuesta inválida. Por favor, elija una opción válida (a, b, c, d).")
            respuesta = input("Elija una opción (a, b, c, d): ").lower()
        respuestas.append(respuesta)
    
    return respuestas

# Calcular el índice de riesgo del cliente basado en las respuestas
def calcular_indice_riesgo(respuestas):
    # Asignar puntos según las respuestas
    puntos = {
        '1': {'a': 2, 'b': 4, 'c': 6, 'd': 10},
        '2': {'a': 2, 'b': 4, 'c': 6, 'd': 10},
        '3': {'a': 10, 'b': 6, 'c': 4, 'd': 2},
        '4': {'a': 2, 'b': 4, 'c': 6, 'd': 10},
        '5': {'a': 2, 'b': 4, 'c': 6, 'd': 10},
        '6': {'a': 2, 'b': 4, 'c': 6, 'd': 10},
        '7': {'a': 4, 'b': 2, 'c': 6, 'd': 10},
        '8': {'a': 2, 'b': 4, 'c': 6, 'd': 10}
    }

    # Sumar los puntos de cada respuesta
    total_puntos = sum(puntos[str(i+1)][respuestas[i]] for i in range(len(respuestas)))
    
    # Calcular el índice de riesgo
    indice_riesgo = total_puntos / len(respuestas)
    
    return indice_riesgo

def clasificar_indice_riesgo(indice_riesgo):
    if 1 <= indice_riesgo <= 2.9:
        return "Muy Conservador"
    elif 3 <= indice_riesgo <= 4.9:
        return "Conservador"
    elif 5 <= indice_riesgo <= 6.9:
        return "Moderado"
    elif 7 <= indice_riesgo <= 8.9:
        return "Agresivo"
    elif 9 <= indice_riesgo <= 10:
        return "Muy Agresivo"

# Obtener las respuestas del cliente
respuestas = obtener_respuestas()

# Calcular el índice de riesgo
indice_riesgo_cliente = calcular_indice_riesgo(respuestas)

# Imprimir el índice de riesgo y su clasificación
print(f"\nEl índice de riesgo es: {indice_riesgo_cliente:.1f}")
print(clasificar_indice_riesgo(indice_riesgo_cliente))

# Escalar el valor de aversión al riesgo
lambda_cliente = indice_riesgo_cliente / 10.0  # Escala a un rango de 0.1 a 1.0
print(f"Valor de aversión al riesgo escalado del cliente: {lambda_cliente}")

```

```python
# Sección 2: Tickers y Black-Litterman

import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from scipy.optimize import minimize

# Obtener la tasa libre de riesgo actual
def obtener_tasa_libre_riesgo():
    # Obtenemos los datos del bono del tesoro de 10 años (^TNX)
    bono_tesoro = yf.Ticker("^TNX")
    tasa_libre_riesgo = bono_tesoro.history(period="1d")['Close'].iloc[-1] / 100
    return tasa_libre_riesgo

# Obtener datos históricos de Yahoo Finance
def obtener_datos_yahoo(tickers, start_date, end_date):
    datos = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
    retornos = datos.pct_change().dropna()
    return retornos

# Calcular retornos esperados usando CAPM
def calcular_retornos_capm(tickers, tasa_libre_riesgo, mercado='^GSPC'):
    end_date = datetime.now()
    start_date = end_date - timedelta(days=5*365)
    
    retornos_activos = obtener_datos_yahoo(tickers, start_date, end_date)
    retornos_mercado = obtener_datos_yahoo([mercado], start_date, end_date)
    
    if isinstance(retornos_mercado, pd.DataFrame):
        retornos_mercado = retornos_mercado.iloc[:, 0]
    
    retornos_activos, retornos_mercado = retornos_activos.align(retornos_mercado, join='inner', axis=0)
    
    betas = []
    retornos_esperados = []
    for ticker in tickers:
        cov_matrix = np.cov(retornos_activos[ticker], retornos_mercado)
        beta = cov_matrix[0, 1] / cov_matrix[1, 1]
        betas.append(beta)
        
        retorno_esperado = tasa_libre_riesgo + beta * (retornos_mercado.mean() * 252 - tasa_libre_riesgo)
        retornos_esperados.append(retorno_esperado)
    
    matriz_covarianza = retornos_activos.cov() * 252  # Covarianza anualizada
    return np.array(retornos_esperados), matriz_covarianza

tickers = input("Ingrese los tickers de los activos separados por comas: ").split(',')

# Obtener la tasa libre de riesgo actual
tasa_libre_riesgo = obtener_tasa_libre_riesgo()
print(f"Tasa libre de riesgo (bonos del tesoro a 10 años): {tasa_libre_riesgo * 100}%")

retornos_esperados_mercado, matriz_covarianza = calcular_retornos_capm(tickers, tasa_libre_riesgo)
print("Retornos esperados calculados por CAPM:", retornos_esperados_mercado)

# Calcular el índice de riesgo para cada activo
def calcular_indice_riesgo_activo(retornos_esperados, matriz_covarianza):
    desviaciones_estandar = np.sqrt(np.diag(matriz_covarianza))
    indice_riesgo = retornos_esperados / desviaciones_estandar
    return indice_riesgo

indice_riesgo_activos = calcular_indice_riesgo_activo(retornos_esperados_mercado, matriz_covarianza)
for ticker, riesgo in zip(tickers, indice_riesgo_activos):
    print(f"Índice de riesgo para {ticker}: {riesgo}")

# Calcular el índice de riesgo del cliente
def calcular_indice_riesgo_cliente(pesos):
    return np.dot(indice_riesgo_activos, pesos)

# Configurar las opiniones del inversor y del cliente
def obtener_opiniones(num_opiniones, num_tickers):
    P = []
    Q = []
    for i in range(num_opiniones):
        while True:
            try:
                opinion = input(f"Ingrese los pesos para la opinión {i+1} (separados por comas): ")
                pesos = list(map(float, opinion.split(',')))
                if len(pesos) != num_tickers:
                    raise ValueError("La cantidad de pesos debe coincidir con la cantidad de tickers.")
                retorno_opinion = input(f"Ingrese el retorno esperado para la opinión {i+1}: ")
                retorno_opinion = retorno_opinion.replace(",", ".")  # Reemplazar coma por punto si es necesario
                retorno_opinion = float(retorno_opinion)
                break
            except ValueError as e:
                print(f"Error: {e}. Por favor, intente nuevamente.")
        P.append(pesos)
        Q.append(retorno_opinion)
    P = np.array(P)
    Q = np.array(Q)
    return P, Q

# Solo una opinión del inversor y del cliente
print("Ingrese la opinión del inversor:")
P_inversor, Q_inversor = obtener_opiniones(1, len(tickers))

print("Ingrese la opinión del cliente:")
P_cliente, Q_cliente = obtener_opiniones(1, len(tickers))

# Combinamos las opiniones del inversor y del cliente
P = np.vstack([P_inversor, P_cliente])
Q = np.concatenate([Q_inversor, Q_cliente])

# Matriz de incertidumbre de las opiniones
omega = np.diag([0.0001] * P.shape[0])

# Calcular π (retornos implícitos del mercado)
pesos_mercado = np.ones(len(tickers)) / len(tickers)
pi = np.dot(matriz_covarianza, pesos_mercado)  # Este valor no se multiplica por lambda_cliente

# Calcular los retornos ajustados con Black-Litterman
M_inverse = np.linalg.inv(np.dot(P.T, np.linalg.inv(omega)).dot(P) + np.linalg.inv(matriz_covarianza))
adjusted_returns = M_inverse.dot(np.dot(P.T, np.linalg.inv(omega)).dot(Q) + np.linalg.inv(matriz_covarianza).dot(pi))

lambda_cliente = 3.0  # Ejemplo: asumiendo un valor fijo o calculado previamente

# Función de utilidad ajustada
def funcion_utilidad(pesos):
    retorno_portafolio = np.dot(pesos, adjusted_returns)
    varianza_portafolio = np.dot(pesos.T, np.dot(matriz_covarianza, pesos))
    utilidad = retorno_portafolio - (lambda_cliente / 2) * varianza_portafolio
    return -utilidad  # Negativo porque vamos a minimizar

# Restricción: la suma de los pesos debe ser 1
def restriccion_suma_pesos(pesos):
    return np.sum(pesos) - 1

# Restricciones de no-negatividad para los pesos, cada peso >= x y respetar el índice de riesgo del cliente
x = float(input("Ingrese el valor mínimo que puede tener cada peso: "))
restricciones = (
    {'type': 'eq', 'fun': restriccion_suma_pesos},
    {'type': 'ineq', 'fun': lambda pesos: pesos - x},  # Todos los pesos deben ser >= x
    {'type': 'ineq', 'fun': lambda pesos: lambda_cliente - calcular_indice_riesgo_cliente(pesos)}  # El índice de riesgo debe ser <= aversión al riesgo del cliente
)

# Pesos iniciales (igual ponderación)
pesos_iniciales = np.ones(len(retornos_esperados_mercado)) / len(retornos_esperados_mercado)

# Optimización
resultado = minimize(funcion_utilidad, pesos_iniciales, method='SLSQP', constraints=restricciones)

# Función para calcular retorno y volatilidad del portafolio
def calcular_retornos_volatilidad_portafolio(pesos, retornos_esperados, matriz_covarianza):
    retorno_portafolio = np.dot(pesos, retornos_esperados)
    volatilidad_portafolio = np.sqrt(np.dot(pesos.T, np.dot(matriz_covarianza, pesos)))
    return retorno_portafolio, volatilidad_portafolio

# Función para calcular el ratio de Sharpe
def calcular_ratio_sharpe(retorno_portafolio, volatilidad_portafolio, tasa_libre_riesgo):
    ratio_sharpe = (retorno_portafolio - tasa_libre_riesgo) / volatilidad_portafolio
    return ratio_sharpe

# Calcular y mostrar el ratio de Sharpe de cada activo
def calcular_sharpe_activos(retornos_esperados, matriz_covarianza, tasa_libre_riesgo):
    desviaciones_estandar = np.sqrt(np.diag(matriz_covarianza))
    ratios_sharpe = (retornos_esperados - tasa_libre_riesgo) / desviaciones_estandar
    return ratios_sharpe

# Mostrar el ratio de Sharpe de cada activo
ratios_sharpe_activos = calcular_sharpe_activos(retornos_esperados_mercado, matriz_covarianza, tasa_libre_riesgo)
for ticker, sharpe_ratio in zip(tickers, ratios_sharpe_activos):
    print(f"Ratio de Sharpe para {ticker}: {sharpe_ratio}")

# Resultados
if resultado.success:
    pesos_optimos = resultado.x
    utilidad_maxima = -resultado.fun
    print("Pesos óptimos del portafolio:", pesos_optimos)
    print("Utilidad máxima:", utilidad_maxima)
    print("Índice de riesgo del portafolio:", calcular_indice_riesgo_cliente(pesos_optimos))
    
    # Calcular y mostrar el ratio de Sharpe del portafolio óptimo
    retorno_portafolio_optimo, volatilidad_portafolio_optimo = calcular_retornos_volatilidad_portafolio(pesos_optimos, adjusted_returns, matriz_covarianza)
    ratio_sharpe_portafolio = calcular_ratio_sharpe(retorno_portafolio_optimo, volatilidad_portafolio_optimo, tasa_libre_riesgo)
    print("Ratio de Sharpe del portafolio óptimo:", ratio_sharpe_portafolio)
else:
    print("La optimización no fue exitosa.")
    print("Mensaje de error:", resultado.message)

# Comparación de Sharpe Ratios
print("\nComparación de Sharpe Ratios:")
for ticker, sharpe_ratio in zip(tickers, ratios_sharpe_activos):
    print(f"Ratio de Sharpe para {ticker}: {sharpe_ratio}")
print(f"Ratio de Sharpe del portafolio óptimo: {ratio_sharpe_portafolio}")

if ratio_sharpe_portafolio > max(ratios_sharpe_activos):
    print("El portafolio óptimo tiene un mejor ratio de Sharpe que cualquier activo individual.")
else:
    print("Algunos activos individuales tienen un mejor ratio de Sharpe que el portafolio óptimo.")
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from datetime import datetime, timedelta

# Obtener los datos históricos y calcular los retornos diarios
def obtener_datos_y_calcular_retornos(tickers, start_date, end_date):
    datos = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
    retornos = np.log(datos / datos.shift(1)).dropna()
    return datos, retornos

# Función para graficar el rendimiento del portafolio optimizado
def graficar_performance_portafolio(tickers, pesos_optimos):
    años = int(input("Ingrese el número de años que desea ver en el gráfico: "))
    end_date = datetime.now()
    start_date = end_date - timedelta(days=años * 365)
    
    precios, retornos_diarios = obtener_datos_y_calcular_retornos(tickers, start_date, end_date)
    
    # Validación de que los datos de precios sean consistentes
    if retornos_diarios.isnull().values.any():
        print("Error: Datos de precios contienen valores nulos.")
        return
    
    # Calcular el balance acumulado del portafolio
    balance_portafolio = (retornos_diarios + 1).cumprod().dot(pesos_optimos)
    
    # Normalizar el balance al primer día para que comience en el mismo punto
    valor_inicial = float(input("Ingrese el valor inicial en dólares para invertir en el portafolio: "))
    normalized_balance = balance_portafolio / balance_portafolio.iloc[0] * valor_inicial
    
    # Función para dar formato a los valores monetarios en el eje Y
    def money_format(x, pos):
        return '${:,.0f}'.format(x)
    
    # Configurar el formato del eje Y
    formatter = FuncFormatter(money_format)
    
    # Graficar el valor del portafolio optimizado
    plt.figure(figsize=(14, 7))
    plt.plot(normalized_balance, label='Black-Litterman', color='red')
    plt.title(f'Performance del Portafolio Optimizado en los Últimos {años} Años')
    plt.xlabel('Fecha')
    plt.ylabel('Valor del Portafolio (USD)')
    plt.gca().yaxis.set_major_formatter(formatter)
    plt.legend()
    plt.grid(True)
    plt.show()


# Graficar los resultados usando los pesos óptimos obtenidos de la optimización
graficar_performance_portafolio(tickers, pesos_optimos)
```
